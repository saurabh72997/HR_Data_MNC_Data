# Q.1) What is the distribution of Employee Status (Active, Resigned, Retired, Terminated) ?

# Q.2) What is the distribution of work modes (On-site, Remote) ?

```python
work=df["Work_Mode"].value_counts().reset_index()
work
```

```python
print(type(work))
```

```python
work.plot(kind='pie', y='count', labels=work['Work_Mode'], autopct='%1.1f%%', explode=(0.03,0.03), shadow=True, figsize=(8, 8), colors=['skyblue', 'lightcoral'])
plt.ylabel('') # Remove the default 'count' label for better aesthetics
plt.title('Distribution of Work Modes')
plt.show()
```

# Q.4) What is the average salary by Department?

```python
dept=df.groupby("Department")["Salary_INR"].mean().astype(int)
dept
```

```python
dept.plot(kind="bar",figsize=(10,6),color=['skyblue','green','red','yellow','black','white','purple','pink'],label=True)
plt.xticks(rotation="vertical")
plt.grid()
plt.title("Average Salary in Department")
plt.ylabel("Salary")
plt.xlabel("Department")
plt.show()
```

#  Q.7) How many employees Resigned & Terminated in each department ?

```python
df.groupby("Department")["Salary_INR"].mean().astype(int).reset_index().sort_values(by="Salary_INR",ascending=False).head(5)
```

```python
df.head(2)
```

```python
df_resigned=df[df["Status"]=="Resigned"].groupby("Department")['Status'].count()
```

```python
df_resigned
```

```python
plt.figure(figsize=(10,6))
df_resigned.plot(x='df_resigned',y=df_resigned.values,kind="bar",color="pink",legend=True,label='Resigned')
plt.grid()
plt.show()
```

```python
df_terminated=df[df["Status"]=="Terminated"].groupby("Department")['Status'].count()
df_terminated
```

```python
plt.figure(figsize=(10,6))
df_terminated.plot(x='df_terminated',y=df_terminated.values,kind="bar",color="green",legend=True,label='Terminated')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
df_resigned.plot(x='df_resigned',y=df_resigned.values,kind="bar",color="pink",legend=True,label='Resigned')
df_terminated.plot(x='df_terminated',y=df_terminated.values,kind="bar",color="black",legend=True,label='Terminated')
plt.grid()
plt.show()
```
